```python
#imported packages
import pandas as pd
import numpy as np
```

```python
#PART ONE: importing the data
```

```python
#resources: https://stackoverflow.com/questions/18022845/how-to-get-set-a-pandas-index-column-title-or-name, 
```

```python
namesfile = pd.read_csv('Popular_Baby_Names.csv')
namesdf  = pd. DataFrame(namesfile)
namesdf.index = np.arange(1, len(namesdf) + 1)
namesdf.index.name = 'Name ID'
print(namesdf)
#sorteddf = namesdf.sort_values('Rank')
#print(sorteddf)
```

```python
#PART TWO: cleaning the data  by doing the following-- checking for empty values and lowercasing all strings
```

```python
#checking for empty values with the isnull() function, will return true for empty values
namesdf.isnull()
#All values are False, so no null values (Will analyze Child's First Name later)
```

```python
#checking for null values again, but will return true if a value is present
namesdf.notnull()
#all values show True, so there are no null values(except for Child's First name which is blank, will get to that next)
```

```python
#Checks for null values in "Child's First Name" since I did not see True or False in previous methods
#Used: https://www.geeksforgeeks.org/working-with-missing-data-in-pandas/

bool_series = pd.isnull(namesdf["Child's First Name"])
namesdf[bool_series]
#No single "Child's First Name box is empty"
```

```python
#Will present the value of the box if it is present
nextbool_series = pd.notnull(namesdf["Child's First Name"])
namesdf[nextbool_series]
#Child's first name has values 
```

```python
#Changing all name strings to lowercase so that data can be queried and return accurate counts
#used: https://sparkbyexamples.com/pandas/convert-pandas-column-to-lowercase/
namesdf["Child's First Name"] = namesdf["Child's First Name"].str.lower()
namesdf['Gender'] = namesdf['Gender'].str.lower()
namesdf['Ethnicity'] = namesdf['Ethnicity'].str.lower()

#, namesdf['Gender'].str.lower(),  namesdf['Ethnicity'].str.lower

print(namesdf)

```

```python
#PART THREE: Querying the data by sorting by popularity,  the instances of a specific name, and printing rows that only meet specific conditions
```

```python
#Ranked names by popularity, by sorting by "Rank" 
namesdf.sort_values('Rank')
```

```python
#Used:https://www.geeksforgeeks.org/how-to-count-occurrences-of-specific-value-in-pandas-column/
#Used .value_counts() to find number of times "Jackson" was counted
#Collectively, from 2011-2019, the name Jackson repeated 83 times- over the course of these years, 83 children were named Jackson in this location 
print(namesdf["Child's First Name"].value_counts()['jackson'])
```

```python

#used https://keytodatascience.com/selecting-rows-conditions-pandas-dataframe/

#Printed the rows that met specific conditions; specically the conditions for hispanic females in 2011 rank first most popular

namesdf[namesdf['Rank'] == 1 & (namesdf['Ethnicity'] == 'hispanic') & (namesdf['Year of Birth'] == 2011) & (namesdf['Gender'] == 'female')]

#My insight from this data is the the most popular name for hispanic females in 2011 in this location was Isabella, counted a total of 331 times
```
